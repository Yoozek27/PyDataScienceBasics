```python
import pandas as pd
import numpy as np

suppliers = pd.read_csv('./data/northwind/suppliers.csv', usecols=['SupplierID','CompanyName','City'])
products = pd.read_csv('./data/northwind/products.csv', usecols=['ProductID','ProductName','SupplierID'])
```

```python
suppliers.merge(products, on='SupplierID').head()
```

```python
suppliers.set_index('SupplierID', inplace=True)
```

```python
suppliers.merge(products, on='SupplierID', left_index=True).head()
```

```python
products.set_index('SupplierID', inplace=True)
```

```python
suppliers.merge(products, left_index=True, right_index=True).head()
```

```python
suppliers.reset_index(inplace=True)
```

```python
suppliers.merge(products, on='SupplierID', right_index=True).head()
```
